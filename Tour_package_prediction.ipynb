# Treating the Missing values

```python
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(value = df['DurationOfPitch'].median())
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(value = df['NumberOfTrips'].median())
df['Age'] = df['Age'].fillna(value = df['Age'].median())
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(value = df['MonthlyIncome'].median())
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(value=df['NumberOfChildrenVisiting'].median())
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()/df.shape[0]*100
```

# Separating Target Column

```python
x = df.drop(['ProdTaken','PitchSatisfactionScore','ProductPitched','NumberOfFollowups','DurationOfPitch'],axis=1)
x = pd.get_dummies(x,drop_first=True)
y = df['ProdTaken']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
print('Training Score:',100*dt.score(x_train,y_train))
print('Testing Score :',100*dt.score(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score

#accuracy_train = accuracy_score(x_train,y_train)

accuracy_test = accuracy_score(y_test, y_pred)

#print('Training accuracy',accuracy_train)

print('Testing accuracy',accuracy_test)
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_prdr = rf.predict(x_test)
```

```python
print('Training Score:', 100 * rf.score(x_train,y_train))
print('Testing Score :', 100 * rf.score(x_test,y_test))
```

```python
accuracy_test = accuracy_score(y_test, y_prdr)
print('Testing accuracy', accuracy_test)
```

# Performance Matrix

```python
from sklearn.metrics import confusion_matrix , classification_report
print('Confusion Matrix:\n ' + str(confusion_matrix(y_prdr,y_test)))
print('_________________________')
print('Classification Report:\n ' + str(classification_report(y_prdr,y_test)))
```

# SVM

```python
from sklearn.svm import SVC
sv_model = SVC()
sv_model.fit(x_train,y_train)
```

```python
y_prds = sv_model.predict(x_test)
```

```python
print('Training Score:',100 * sv_model.score(x_train,y_train))
print('Testing Score :',100 * sv_model.score(x_test,y_test))
```

```python
accuracy_test = accuracy_score(y_test, y_prds)
print('Testing accuracy', accuracy_test)
```

```python
print('Confusion Matrix:\n ' + str(confusion_matrix(y_prds,y_test)))
print('_________________________')
print('Classification Report:\n ' + str(classification_report(y_prds,y_test)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train,y_train)
```
